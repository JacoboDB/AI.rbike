```python
chicagodf['estacion inicio'] = [', '.join(str(x) for x in y) for y in map(tuple, chicagodf[['latitude_start', 'longitude_start']].values)]
chicagodf.head()
```

```python
def get_distance(lat1,long1, lat2, long2):
  #key from https://developer.mapquest.com/documentation/samples/directions/v2/route/
  r = requests.get('https://www.mapquestapi.com/directions/v2/route?key=lYrP4vF3Uk5zgTiGGuEzQGwGIVDGuy24&unit=k&from='+str(lat1)+'%2C'+str(long1)+'&to='+str(lat2)+'%2C'+str(long2)+'&outFormat=json&ambiguities=ignore&routeType=bicycle&doReverseGeocode=false&enhancedNarrative=false&avoidTimedConditions=false')
  return r.json()['route']['distance']
```

```python
chicago1df['Distancia Recorrida'] = chicago1df.apply(lambda x: get_distance(x['latitude_start'],x['longitude_start'], x['latitude_end'],x['longitude_end']),axis=1)
```
